```python
import networkx as nx
from IPython.display import display
from ipywidgets import Dropdown
get_ipython().magic('matplotlib inline')
```

```python
import tega.driver
d = tega.driver.Driver(host='192.168.57.133')
d.rpc('plugins.hook')
```

```python
def on_value_change(name, value):
    d.rpc('plugins.server_client', port=int(value))
    print(value)
    server_client = d.get(path='graph.server_client.{}'.format(value))
    g = nx.DiGraph(server_client)
    nx.draw_spring(g, node_size=600, with_labels=True, arrows=True, alpha=0.8)
```

```python
port_list = ['---port---', '179', '8739']
dw = Dropdown(description='port_list', options=port_list)
display(dw)
dw.on_trait_change(on_value_change, 'value')
```
